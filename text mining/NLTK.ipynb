```python
import nltk
import konlpy
from nltk import word_tokenize
from collections import Counter
```

```python
sentence='''절망의 반대가 희망은 아니다.
어두운 밤하늘에 별이 빛나듯
희망은 절망 속에 싹트는 거지
만약에 우리가 희망함이 적다면
그 누가 세상을 비추어줄까.
정희성, 희망 공부'''
```

```python
from konlpy.tag import Okt
t= Okt()
```

```python
print('형태소:', t.morphs(sentence))
print()
print('명사:', t.nouns(sentence))
print()
print('품사 태깅 결과:', t.pos(sentence))
```

```python
from konlpy.corpus import kolaw
const_doc = kolaw.open('constitution.txt').read()

print(type(const_doc))
print(len(const_doc))
print(const_doc[:600])
```

```python
t = Okt()
tokens_const = t.morphs(const_doc)

print('#토큰의 수:', len(tokens_const))
print('#앞 100개의 토큰')
print(tokens_const[:100])
```

```python
tokens_const = t.nouns(const_doc)  #형태소 단위로 tokenize 후 명사만 추출
print('#토큰의 수:', len(tokens_const))
print('#앞 100개의 토큰')
print(tokens_const[:100])
```

```python
tokens_const = [token for token in tokens_const if len(token)  > 1] #한 단어인 명사는 제외 (애, 것, 함, 제, 장)
print('#토큰의 수:', len(tokens_const))
print('#앞 100개의 토큰')
print(tokens_const[:100])
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

const_cnt = {}
for word in tokens_const:
    const_cnt[word] = const_cnt.get(word,0) + 1
    
def word_graph(cnt, max_words=10):
    sorted_w = sorted(cnt.items(), key=lambda kv: kv[1])
    print(sorted_w[-max_words:])
    n, w = zip(*sorted_w[-max_words:])
    
    plt.barh(range(len(n)), w,tick_label=n)
    plt.show()
    
word_graph(const_cnt, max_words=20)
```

```python
categories = ["카테고리1", "카테고리2"]
risk_scores = {"카테고리1": 0.3, "카테고리2": 0.5}
category_weights = {
    1: 1.0,  # 1개 카테고리에 속한 경우의 가중치
    2: 1.2,  # 2개 카테고리에 속한 경우의 가중치
    3: 1.5   # 3개 카테고리에 속한 경우의 가중치
}

# 항목이 속한 카테고리 목록
item_categories = ["카테고리1", "카테고리2", "카테고리3"]

# 속한 카테고리의 개수 계산
category_count = len(item_categories)

# 가중치 할당
if category_count in category_weights:
    weight = category_weights[category_count]
else:
    # 원하는 방식으로 처리 (예: 특별한 경우에 대한 가중치 설정)
    weight = 1.0

# 위험도 계산
risk_score = 0.3  # 카테고리의 위험도 (예시)
total_risk = risk_score * weight

print("총 위험도:", total_risk)
```

```python
# 단어별 빈도수 계산
word_counts = Counter(tokens_const)

# 가장 빈도가 높은 단어를 찾음
most_common_word, most_common_count = word_counts.most_common(1)[0]

# 결과 출력
print(f"가장 빈도가 높은 단어: {most_common_word}, 빈도수: {most_common_count}")
```
